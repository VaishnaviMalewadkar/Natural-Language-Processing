```python
l = ['description', 'company_profile', 'requirements', 'employment_type', 'location','benefits']
for i in l:
    jp[i] = jp[i].apply(str)  
```

```python
jp.rename(columns={'fraudulent':'Y'},inplace=True)
```

```python
jp=jp.loc[:,['X','Y']]
jp.head()
```

```python
jp.X=jp.X.str.lower() # before builing any NLP model convert it into either lower or uppercase
jp.head()
```

```python
def text_process(mess): #creating a function
    """"
    1. remove the punctuation
    2. remove the stopwords
    3. return the list of clean textwords
    """
    global nopunc
    nopunc=[char for char in mess if char not in string.punctuation]
    nopunc="".join(nopunc)
    l1=list(stopwords.words('English'))
    return [word for word in nopunc.split() if word not in l1 ]
```

```python
jp.isnull().sum()[jp.isnull().sum()>0]
```

```python
jp.shape
```

```python
import string
from nltk.corpus import stopwords
```

```python
from sklearn.feature_extraction.text import CountVectorizer
bow_transformer= CountVectorizer(analyzer=text_process).fit(jp['X'])
```

```python
TDM=bow_transformer.transform(jp['X'])
```

```python
TDM.shape
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
from sklearn.model_selection import train_test_split

jp_train_x, jp_test_x, jp_train_y,jp_test_y = train_test_split(TDM ,jp.Y)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
logreg_jp=LogisticRegression()
logreg_jp.fit(jp_train_x,jp_train_y)
```

```python
pred_log_jp=logreg_jp.predict(jp_test_x)
```

```python
confusion_matrix(jp_test_y,pred_log_jp)
```

```python
print(classification_report(jp_test_y,pred_log_jp))
```

```python
from sklearn.tree import DecisionTreeClassifier

dt_jp=DecisionTreeClassifier(class_weight='balanced')
dt_jp.fit(jp_train_x,jp_train_y)
```

```python
pred_dt_jp=dt_jp.predict(jp_test_x)
```

```python
confusion_matrix(jp_test_y,pred_dt_jp)
```

```python
print(classification_report(jp_test_y,pred_dt_jp))
```
